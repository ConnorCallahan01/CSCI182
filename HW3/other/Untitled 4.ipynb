```python
#The purpose of this block of code was to create an empty dictionary with the key being the order_id, and the values
#   being the empty lists that I will later fill with each product purchased during that order

orders = [] #initiate an empty list
for value in olist_order_items['order_id']:
    orders.append(value) #add the order_id to the list

orders
orders_dict = {} #initiate the dictionary
for items in orders:
        orders_dict[items] = [] #make the values of each order an empty list that will later be filled
#print the dictionary to prove it worked: 
orders_dict
```

```python
#This block of code now adds the english product name to the dictionary. I iterated through each order_id in the 
#  olist_order_items dataframe and checked if it was equal to a key in the dictionary. If it was, I appened the 
#  english product name to the key-value list. 
i = 0
for value in olist_order_items['order_id']: #iterate through order_ids
    for key in orders_dict: #iterate through keys in 
        if key == value: #if the value is equal to the key in the dictionary
            orders_dict[key].append(olist_order_items.loc[i, "prod_category_name_english"]) #add that value to the key-value list. 
            break
    i += 1


```

```python
# For this block of code, I turned the dictionary into a dataframe where the rows are each order_id and the columns 
#  are the item that was bought during the transaction. 
orderId_producst_df = pd.DataFrame.from_dict(orders_dict, orient='index')
orderId_producst_df = orderId_producst_df.fillna(0) #this was to fill any NaN or None value to 0 so I could know if there was missing data or not. 

# Print the dataframe to show that it worked: 
orderId_producst_df

```

```python
# This next part I modeled off of the sample that was given to us. 
# The apriori function that we are going to use requires a list of lists. To do this I created an empty list for that purpose that will be appended to. 
records = []
for i in range(0, 98666): #For all the rows in the dataframe
    records.append([str(orderId_producst_df.values[i,j]) for j in range(0, 21)]) 
    
    
```

```python
#This block of code finds the association_rules for the Market-Basket Analysis. 
association_rules = apriori(records, min_support=0.0000015, min_confidence=0.00005, min_lift=1.1)
association_results = list(association_rules)
```

- 2. Find how many tokens (words and punctuation symbols) are in the text. A token is a linguistic unit such as a word, punctuation mark, or alpha-numeric strings.
- 3. Find how many of the tokens found in 1.2) are unique.
- 4. Find how many tokens are unique after removing stopwords.
- 5. What is the lexical diversity of the given text input? (i.e. ratio of unique tokens to the total number of tokens)

```python
import nltk

import re
import pprint
from nltk import word_tokenize
from urllib import request

# 1: Import the libraries required and Set up Data. Use the link ”https://www.gutenberg.org/files/2701/old/moby10b.txt” to access the ebook.
# 2: Find how many tokens (words and punctuation symbols) are in the text. A token is a linguistic unit such as a word, punctuation mark, or alpha-numeric strings.

url = "https://www.gutenberg.org/files/2701/old/moby10b.txt" # specify the URL
response = request.urlopen(url) #open the URL
raw = response.read().decode('utf-8-sig') #access the text and read it into variable RAW
tokens = word_tokenize(raw) #tokenize each value 
print(tokens[:10]) #print the beginning to show some of the tokens
print(len(tokens)) #print the length of the tokens. 
```

```python
# 3: Find how many of the tokens found in 1.2) are unique.
fdist1 = nltk.FreqDist(tokens)
print("\u0332".join('Before stop words: ')
print("The amount of unique words in the page is: ", len(fdist1))
# print(fdist1.most_common())

# 4: Find how many tokens are unique after removing stopwords.
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
filtered_words = []
for w in tokens:  
    if w not in stop_words and w != '*' and w != '#':
        filtered_words.append(w)
        
fdist2 = nltk.FreqDist(filtered_words)
print("The amount of unique words in the page is: ", len(fdist2))
```

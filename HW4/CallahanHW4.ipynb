```python
import re
source = '/Users/connorcallahan/Desktop/CSCI182 Web Mining/HW4'
allfiles = os.listdir(source)

#this was a function I created to open all the .txt files in the source directory and add them to the temp list
def openDoc(source_file):
    temp = [] #initiate empty array
    for f in source_file: #for all files in source_file
        if f.endswith(".txt"): #if it's one of the .txt files
            x = open(f, 'r', encoding="ISO-8859-1") #open it and read it
            temp.append(x) #add it's contents as a list in the temp list
    return temp #return the list of lists. 
    
text_files = openDoc(allfiles) #call the function for all documents in the source directory
list_of_files = [] 
for i in text_files: #for each document
    g = i.read() #read the contents
    large_set_of_words = g.split(' ') #split the string by space
    list_of_files.append(large_set_of_words) #put all the newly split strings in the list_of_files list I created. 



cleaned_list_temp = []
for i in list_of_files: #for each list in the most recent list
    temp_list = [] #make another temporary list
    for j in i: #for each string in the sublist
        a_string = j; #set it to a string
        alphanumeric = " " #initiate empty string
        for character in a_string: #for each character in the string
            if character.isalnum(): #check to see if it's either a number or letter
                alphanumeric += character #if it only a letter or number, the character is added to the string
            else: 
                alphanumeric += " " #otherwise, a space is add to the string
        alphanumeric = alphanumeric.lower() #make the new strings all lowercase
        temp_list.append(alphanumeric) #add it to the new temp list
    #this was simply to observe take care of the extra space issue I had. It appeared that because I added a space if 
    #the value was neither number or string, an extra, unwanted space was in the string. This got rid of that. 
    s = "".join(temp_list)
    cleaned_list_temp.append(s)


cleaned_list_real = []
for i in cleaned_list_temp: 
    bagOfWordsB = i.split(' ')
    cleaned_list_real.append(bagOfWordsB)


print(cleaned_list_real)
```

```python
#remove all stop words from the documents
from nltk.corpus import stopwords
engl_stopwords = stopwords.words('english')


#first take out all the extra spaces in each list, we don't want to count these as words
cleaned_list_real_2 = []
for list_ in cleaned_list_real:
    res = []
    for ele in list_:
        if ele.strip():
            res.append(ele)
    cleaned_list_real_2.append(res)


#before removing the stop words, we have to calculate and store the total number of words in each document for later.    
document_wl = []
for list_ in cleaned_list_real_2:
    document_wl.append(len(list_))
print(document_wl)   

#this piece of code will make a dictionary of all the words in the document and the count of that word in the doc. 
# it also takes out the stop words in the process. 
keyw_dict_list = [] #initialize the list
for i in cleaned_list_real_2: #for each document
    temp_dict = {} #initialize a dictionary
    for j in i: #for values in each document
        if j not in engl_stopwords: #if the word is not a stop word
            if j in temp_dict: #if it's already in the dictionary
                temp_dict[j] += 1 #make the word count for that word go up
            else: 
                temp_dict[j] = 1 #else addthe word in the document
    keyw_dict_list.append(temp_dict) #add the dictionary to the list so it is a list of dictionaries

for i in keyw_dict_list:
    print(i)
    print()
```

```python
import pandas as pd

#now we will make a dataframe out of the dictionaries to see the word counts of each word in each document 
df = pd.DataFrame(keyw_dict_list) #this will make a DataFrame with the columns being the words and the rows the documents
#There are naturally NaN values if the word is not in the document
#To deal with those values we will make those NaN values be 0 so we can continue with calculation without errors. 
for i in df:
    df[i] = df[i].fillna(0)
    
#make the last column the sum of words in each document
df['Doc Word Sum'] = document_wl

#Now we sum the amount of times the word shows up in all the documents    
column_sum = {} #create a dictionary
for column in df: #for each word
    sum_col = df[column].sum() #find the column sum
    column_sum[column] = sum_col
    
#this makes the column_sum dictionary a new row
print(len(column_sum))
df = df.append(column_sum, ignore_index = True)
tf = df
df
```

```python
#find the tf = term_frequency = number of times a given term appears in document = t/#t in d

#for each word in the dataframe not including the doc_sum_count column
for column in tf.iloc[:,:-1]:
    count = 0 #initialize a counter
    temp_list_tf = [] #inlitialize a tf list
    #for each row in the dataframe
    for row in tf[column]:
        
        denom = tf.loc[count,'Doc Word Sum'] #set the denominator as the word count of the document
        tf_value = row/denom #divide the cell value by the denominator value
        temp_list_tf.append(tf_value) #add that new value to the list
        count +=1 #the counter represents the row so make that go up
        if count == 15: #if the index is at the last row
            temp_list_tf.append(column_sum[column])
            break #break the loop
    tf[column] = temp_list_tf #reset the value of the column to the tf value

tf
```

```python
# find the document frequency = amount of times the word is in the document / amount of documents
dfreq = pd.DataFrame(keyw_dict_list)
for i in dfreq:
    dfreq[i] = dfreq[i].fillna(0)
    
temp_list = []
for column in dfreq: 
    counter = 0
    for row in dfreq[column]:
        if row != 0:
            counter += 1
    temp_list.append(counter)

d_freq = {}
counter = 0
for column in dfreq:
    for i in temp_list:
        #print(temp_list[counter])
        d_freq[column] = temp_list[counter]/len(dfreq.index)
        counter +=1
        
        break

dfreq = dfreq.append(d_freq, ignore_index = True)
dfreq
```

```python
# calculating idf: log(N/(df+1))
import math
term_idf = {}
for term in d_freq: 
    df_and_1 = d_freq[term] +1
    n_div_df1 = (len(dfreq.index)-1) / df_and_1
    idf = math.log2(n_div_df1)
    term_idf[term] = idf
    
term_idf
```

```python
# Final step: TF*IDF --> TF-IDF: 
for column in tf.iloc[:,:-1]: 
    tf[column] = tf[column] * term_idf[column]
    
tf
```

```python
tdIDF = tf.to_dict('index')
```
